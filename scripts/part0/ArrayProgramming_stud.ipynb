```julia
# 1D Arrays: define an axis x ranging from -1.0 to 1.0 with 100 points 
x = LinRange( -1.0, 1.0, 100 )
```

```julia
# 1D Arrays: Evaluate a Gaussian distribution, f: amplitude (A) is 1.0, standard deviation (σ) is 0.2 
A = 1.0
σ = 0.2
f = A  * exp.( -0.5 * ( x ./ σ ).^2 )
```

```julia
# 1D Arrays: Visualise a Gaussian distribution, f 
plot( x, f, xlabel="x", ylabel="y", title="My first Gaussian", label="f(x)" )
```

```julia
# 1D Arrays: Create a copy of the Gaussian, g
g = copy(f)
```

```julia
# 1D Arrays: add 1.0 to the first half of g
g[1:50] .+= 1.0
```

```julia
# 1D Arrays: multiply second half of g by 2.0
g[51:end] .*= 2.0
```

```julia
# 1D Arrays: plot f(x) as line plot and g(x) as a scatter plot 
plot( x, f, xlabel="x", ylabel="y", title="My first Gaussians", label="f(x)" )
scatter!( x, g, label="g(x)" )
```

## 2D

```julia
# Generate a 2D model space, and an y axis both ranging from -1.0 to 1.0 with 100 nodes
x = LinRange( -1.0, 1.0, 100 )
y = LinRange( -1.0, 1.0, 100 )
```

```julia
# Create and visualise a 2D Gaussian, h(x,y) 
h = A  * exp.( -0.5 * ( x ./ σ ).^2 )' .* exp.( -0.5 * ( y ./ σ ).^2 )
```

```julia
# Visualise using a heatmap
heatmap(x, y, h, xlabel="x", ylabel="y", title="2D Gaussian", aspect_ratio=1., xlim=(-1,1))
```

```julia
# Visualise using a contour plot
contour(x, y, h, xlabel="x", ylabel="y", title="2D Gaussian", aspect_ratio=1., xlim=(-1,1))
```

```julia
# Visualise using a contourf
contourf(x, y, h, xlabel="x", ylabel="y", title="2D Gaussian", aspect_ratio=1., xlim=(-1,1))
```
