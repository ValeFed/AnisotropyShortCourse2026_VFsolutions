```julia
# Load AD package
using ForwardDiff
```

### The derivative

```julia
# Function: scalar --> scalar
f(x) = x^3 + 2x

# Input: scalar
x = 2.0

# Derivative: scalar --> scalar --> scalar
ForwardDiff.derivative(f, 2.0)
```

### The gradient

```julia
# Function: vector --> scalar
g(x) = x[1]^2 + x[2]^2

# Input: vector
x = [3.0, 4.0]

# Gradient: vector --> scalar --> vector
ForwardDiff.gradient(g, x)
```

### The Jacobian

```julia
# Function: vector --> vector
h(x) = [
    x[1]^2,
    x[1]*x[2]
]

# Input: vector
x = [2.0, 3.0]

# Jacobian: vector --> vector --> matrix
ForwardDiff.jacobian(h, x)
```

### The Hessian

```julia
# Function: vector --> scalar
i(x) = x[1]^2 + 3*x[1]*x[2] + x[2]^2

# Input: vector
x = [1.0, 2.0]

# Hessian: vector --> scalar --> matrix
ForwardDiff.hessian(i, x)
```
